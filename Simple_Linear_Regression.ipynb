```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
house_price={
    'area':[2600,3000,3200,3600,4000],
    'price':[500000,550000,610000,680000,725000]
}

df=pd.DataFrame(house_price)
df
```

```python
%matplotlib inline
plt.scatter(df.area,df.price,color="red",marker="+")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
168.49315068*3300+60342.46575342456
```

```python
area={
    'area':[3300,3120,3340,5214,1245]
}
d=pd.DataFrame(area)
d
```

```python
p=reg.predict(d)
d['price']=p
d
```

```python
d.to_csv("Predicted.csv",index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
house_price={
    'area':[2600,3000,3200,3600,4000],
    'price':[500000,550000,610000,680000,725000]
}

df=pd.DataFrame(house_price)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```
