```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv("StudentsPerformance (1).csv")
df
```

```python
df.isna().sum()
```

```python
df['gender']=LabelEncoder().fit_transform(df['gender'])
df['parental level of education']=LabelEncoder().fit_transform(df['parental level of education'])
df['test preparation course']=LabelEncoder().fit_transform(df['test preparation course'])
```

```python
X=df.drop(['race/ethnicity','lunch'],axis="columns")
```

```python
df['average_score']=(df['math score']+df['reading score']+df['writing score'])/3
```

```python
df
```

```python
df['pass']=df['average_score'].apply(lambda x:1 if x>40 else 0)
```

```python
df
```

```python
y=df['pass']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```
